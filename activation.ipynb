```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------- Activation Functions ---------------- #
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x) ** 2

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def elu_derivative(x, alpha=1.0):
    return np.where(x > 0, 1, alpha * np.exp(x))

def softmax(x):
    e_x = np.exp(x - np.max(x))  # stability trick
    return e_x / np.sum(e_x, axis=0)

# For simplicity, derivative shown only for single variable case
def softmax_derivative(x):
    s = softmax(x)
    return s * (1 - s)   # valid only for single-variable plotting


# ---------------- Plotting ---------------- #
x = np.linspace(-5, 5, 500)

functions = {
    "Sigmoid": (sigmoid, sigmoid_derivative),
    "Tanh": (tanh, tanh_derivative),
    "ReLU": (relu, relu_derivative),
    "ELU": (elu, elu_derivative),
    "Softmax": (softmax, softmax_derivative)
}

fig, axes = plt.subplots(len(functions), 2, figsize=(10, 15))

for i, (name, (func, dfunc)) in enumerate(functions.items()):
    y = func(x)
    dy = dfunc(x)

    axes[i, 0].plot(x, y, label=f"{name}", color="blue")
    axes[i, 0].set_title(f"{name} Function")
    axes[i, 0].grid(True)
    axes[i, 0].legend()

    axes[i, 1].plot(x, dy, label=f"{name} Derivative", color="red")
    axes[i, 1].set_title(f"{name} Derivative")
    axes[i, 1].grid(True)
    axes[i, 1].legend()

plt.tight_layout()
plt.show()
```
