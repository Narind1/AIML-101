```python
import numpy as np


# Simple Neural Network from Scratch (1 hidden layer)
class NeuralNetwork:
def __init__(self, input_size, hidden_size, output_size, learning_rate=0.1):
# Initialize weights
self.W1 = np.random.randn(input_size, hidden_size) * 0.01
self.b1 = np.zeros((1, hidden_size))
self.W2 = np.random.randn(hidden_size, output_size) * 0.01
self.b2 = np.zeros((1, output_size))
self.lr = learning_rate


def sigmoid(self, z):
return 1 / (1 + np.exp(-z))


def sigmoid_derivative(self, z):
return z * (1 - z)


def forward(self, X):
# Hidden layer
self.z1 = np.dot(X, self.W1) + self.b1
self.a1 = self.sigmoid(self.z1)
# Output layer
self.z2 = np.dot(self.a1, self.W2) + self.b2
self.a2 = self.sigmoid(self.z2)
return self.a2


def backward(self, X, y, output):
m = y.shape[0]


# Output layer error
error_output = output - y
dW2 = (1/m) * np.dot(self.a1.T, error_output * self.sigmoid_derivative(output))
db2 = (1/m) * np.sum(error_output * self.sigmoid_derivative(output), axis=0, keepdims=True)


# Hidden layer error
error_hidden = np.dot(error_output * self.sigmoid_derivative(output), self.W2.T)
dW1 = (1/m) * np.dot(X.T, error_hidden * self.sigmoid_derivative(self.a1))
db1 = (1/m) * np.sum(error_hidden * self.sigmoid_derivative(self.a1), axis=0, keepdims=True)


# Update weights
self.W2 -= self.lr * dW2
self.b2 -= self.lr * db2
self.W1 -= self.lr * dW1
self.b1 -= self.lr * db1


def train(self, X, y, epochs=10000):
for i in range(epochs):
output = self.forward(X)
self.backward(X, y, output)
if i % 1000 == 0:
loss = np.mean(np.square(y - output))
print(f"Epoch {i}, Loss: {loss}")


# Example: XOR problem
if __name__ == "__main__":
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])


nn = NeuralNetwork(input_size=2, hidden_size=4, output_size=1, learning_rate=0.1)
nn.train(X, y, epochs=10000)


print("Final predictions:")
print(nn.forward(X))
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
def sigmoid_derivative(z):
    return z*(1-z)
```

```python
y= y.reshape(-1, 1)
for i in range(epochs):
    # Forward pass
    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1, W2) + b2
    a2 = sigmoid(z2)

    # Backward pass
    error_output = a2 - y
    dW2 = np.dot(a1.T, error_output * sigmoid_derivative(a2)) / X.shape[0]
    db2 = np.sum(error_output * sigmoid_derivative(a2), axis=0, keepdims=True) / X.shape[0]
    
    error_hidden = np.dot(error_output * sigmoid_derivative(a2), W2.T)
    dW1 = np.dot(X.T, error_hidden * sigmoid_derivative(a1)) / X.shape[0]
    db1 = np.sum(error_hidden * sigmoid_derivative(a1), axis=0, keepdims=True) / X.shape[0]
    
    # Update weights
    W2 -= lr * dW2
    b2 -= lr * db2
    W1 -= lr * dW1
    b1 -= lr * db1
    
    if i % 1000 == 0:
        loss = np.mean((y - a2) ** 2)
        print(f"Epoch {i}, Loss: {loss}")

print("Final predictions:")
print(a2)
```
