```python
import pypandoc

# Input Word file
input_file = "FIRE SAFETY AND PROTECTION NARIND VERMA.docx"

# Output LaTeX file
output_file = "springer_paper.tex"

# Springer Nature LaTeX class
springer_template = r"""
\documentclass[sn-basic]{sn-jnl}  % Springer Nature template class
\usepackage{graphicx}
\usepackage{amsmath}
\usepackage{hyperref}
\usepackage{lineno}
\begin{document}
$body$
\end{document}
"""

# Convert DOCX → LaTeX (body only)
latex_body = pypandoc.convert_file(input_file, 'latex', format='docx')

# Combine with Springer template
with open(output_file, 'w', encoding='utf-8') as f:
    f.write(springer_template.replace('$body$', latex_body))

print(f"✅ Converted successfully! File saved as {output_file}")
```
