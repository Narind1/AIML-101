```python
import numpy as np
X= np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])
```

```python
def step(x):
    return 1 if x>=0 else 0
```

```python
epochs = 10
lr = 0.01
weights = np.zeros(X.shape[1])
bias =0 
```

```python
for epoch in range(epochs):
    for xi, target in zip(X,y):
        op = np.dot(xi,weights)+bias
        y_pred = step(op)
        update = lr * (target-y_pred)
        weights += update *xi
        bias += update
    print("Weights:",weights)
    print("Bias:",bias)
```
